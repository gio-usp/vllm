```python
from transformers import pipeline

pipe = pipeline("text-generation", model="mistralai/Mistral-Nemo-Base-2407")    # Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-Nemo-Base-2407")
model = AutoModelForCausalLM.from_pretrained("/content/drive/Shareddrives/sft/models/mistral-nemo")
```

```python
inputs = tokenizer("the sky is ", return_tensors="pt")
outputs = model.generate(**inputs, max_new_tokens=20)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
